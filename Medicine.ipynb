```python
folder1 = 'C:/Users/lenovo/Desktop/Clinical_HL7_Samples/ELR-ExampleMessages-1.9.2-012216/Elr-ExampleMessages-1.9.2-012216'
folder2 = 'C:/Users/lenovo/Desktop/Clinical_HL7_Samples/Immunization-ExampleMessages-1.9-021214'
folder3 = 'C:/Users/lenovo/Desktop/Clinical_HL7_Samples/Older'
folder4 = 'C:/Users/lenovo/Desktop/Clinical_HL7_Samples/Syndromic-ExampleMessages-1.6-081313/Syndromic-ExampleMessages-1.6-081313'
```

```python
lst = []
folders = [folder1, folder2, folder3,folder4]
for n in folders:
    for root,d_names,f_names in os.walk(n):
        for i in f_names:
            with open(n+'/'+i, 'r') as f:
                for seg in f:
                    if seg.startswith('PID')==True:
                        fields = seg.split('|')
                        lst.append(fields)

df = pd.DataFrame(lst)
```
